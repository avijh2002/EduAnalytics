```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("dataset.csv")
print(df.head)
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

<h1 style="font-size:25px;">DROP UNNAMED COLUMN (5-Oct --> 5-10)</h1>

```python
df=df.drop("Unnamed: 0",axis=1)
```

```python
print(df.head())
```

```python
df.head()
```

<h1 style="font-size:50px;">-->Gender Distribution</h1>


```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

<h1 style="font-size:30px; color:blue;">Count of female students is slighly higher than count of male students</h1>


<h1 style="font-size:50px;">-->analyzing correlation of parents education with students performance</h1>

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(3,3))
sns.heatmap(gb,annot=True)
plt.show()
```

<h1 style="font-size:30px; color:blue;">Students whose parents have a mastes degree tend to have good performance</h1>




<h1 style="font-size:50px;">-->analyzing correlation of parents' maritial status with students performance</h1>

```python
gb=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot=True)
plt.show()
```

<h1 style="font-size:30px; color:blue;">According to the chart there is negligible difference in mean score of student based on maritial status of the parents</h1>







<h1 style="font-size:50px;">-->Analysis of correlation of performance of student with how often does he/she practices sports </h1>

```python
gb=df.groupby("PracticeSport").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot=True)
plt.show()
```

<h1 style="font-size:30px; color:blue;">The above data suggests that students that play sports have a slight edge in performance as compared to thos that dont</h1>



<h1 style="font-size:50px;">-->analyzing correlation of whether student is a first child , to his/her performance</h1>

```python
gb=df.groupby("IsFirstChild").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot=True)
plt.show()
```

<h1 style="font-size:30px; color:blue;">-->According to the chart there is negligible difference in mean score of students based on whether they are the first child to their parents or not</h1>

<h1 style="font-size:50px;">-->Correlation between number of siblings of student to his performance</h1>

```python
df['AverageScore'] = df[['MathScore', 'ReadingScore', 'WritingScore']].mean(axis=1)
grouped_df = df.groupby('NrSiblings')['AverageScore'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(grouped_df['NrSiblings'], grouped_df['AverageScore'])
plt.xlabel('Number of Siblings')
plt.ylabel('Average Score')
plt.title('Relationship Between Number of Siblings and Average Score')
plt.show()
```

<h1 style="font-size:30px; color:blue;">No noteworthy correlation of number of siblings of the student with his performance</h1>


<h1 style="font-size:50px;">-->Box Plot Analysis</h1>

```python
sns.boxplot(data=df,x="MathScore")
plt.show()
sns.boxplot(data=df,x="ReadingScore")
plt.show()
sns.boxplot(data=df,x="WritingScore")
plt.show()
```

<h1 style="font-size:30px; color:blue;">The above data suggests that students face the most difficulty in the subject of maths, with mean value only around 20</h1>



<h1 style="font-size:50px;">-->Distribution Of Ethnic Groups</h1>

```python
print(df["EthnicGroup"].unique())
```

```python
a=df.loc[(df["EthnicGroup"]=='group A')].count()
b=df.loc[(df["EthnicGroup"]=='group B')].count()
c=df.loc[(df["EthnicGroup"]=='group C')].count()
d=df.loc[(df["EthnicGroup"]=='group D')].count()
e=df.loc[(df["EthnicGroup"]=='group E')].count()

l=["group A","group B","group C","group D","group E"]
mlist=[a["EthnicGroup"],b["EthnicGroup"],c["EthnicGroup"],d["EthnicGroup"],e["EthnicGroup"]]
plt.pie(mlist, labels=l, autopct= '%1.2f%%')
plt.show()
```

<h1 style="font-size:30px; color:blue;">The above data suggests that distribution of ethnic group C is highest whereas that of ethnic group A is lowest</h1>
